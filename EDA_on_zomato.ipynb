```python
missPercent=zomato_dataset_copy.isnull().sum() * 100 / len(zomato_dataset_copy)
```

```python
# Renaming columns
zomato_dataset_copy.rename({'Has Table booking':'Accepts_Table_Booking',
                            'Rating color':'Rating_Schema',
                            'Rating text':'Rating',
                            'Average Cost for two':'Avg_Cost',
                            'Has Online delivery':'Online_Delivery',
                            'Aggregate rating':'Aggregate_Rating'},
                           axis=1,
                           inplace=True)
```

```python
# Taking subset from dataset
zomato_dataset_copy_subset=zomato_dataset_copy[['Aggregate_Rating','Rating_Schema','Rating','Votes']]
```
